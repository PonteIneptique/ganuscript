```python
import requests
import requests_cache
import csv

requests_cache.install_cache()


def page_int(label):
    return "".join([c for c in label if c.isnumeric()])


def get_start_end(manifest):# -> Tuple[int, int]
    start, end = None, None
    for idx, page in enumerate(manifest["sequences"][0]["canvases"]):
        if not start and page["label"][0].isnumeric():
            start = page_int(page["label"])
        elif start and page["label"][0].isnumeric():
            end = page_int(page["label"])
        else:
            print(page["label"])
    print(start, end, idx)
    if start and end:
        return start, end
    elif start:
        return start, len(manifest["sequences"][0]["canvases"])
    else:
        return 3, len(manifest["sequences"][0]["canvases"])
    
# Read input CSV
out = []

# Read input CSV
with open("Ganuscrit - MSS.csv") as f:
    reader = csv.DictReader(f)
    
    for row in reader:
        # Compute the manifest if necessary
        if "{ID}" in row["Manifest"]:
            row["Manifest"] = row["Manifest"].format(ID=row["ID"])
    
        # Get start and end
        if not row["start"] and row["Manifest"]:
            d = requests.get(row["Manifest"]).json()
            row["start"], row["end"] = get_start_end(d)
            
        # Get an id if we got the manifest
        if not row["ID"]:
            if row["Source"] == "gallica":
                row["ID"] = row["Manifest"].replace(
                    "/manifest.json", "").replace(
                    "https://gallica.bnf.fr/iiif/ark:/12148/", "")
            elif row["Source"] == "e-codices":
                row["ID"] = row["Manifest"].replace(
                    "https://www.e-codices.unifr.ch/metadata/iiif/fmb-", "").replace(
                    "/manifest.json", "")
        out.append(row)

# Write to output
with open("ganuscrit.csv", "w") as f:
    w = csv.DictWriter(f, fieldnames=list(row.keys()))
    w.writeheader()
    w.writerows(out)
```
